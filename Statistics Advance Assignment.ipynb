Q1. What is the Probability Density Function (PDF)?
   - The Probability Density Function, or PDF, is a way to describe how likely different values are in a continuous         probability distribution. It shows the likelihood of a random variable taking on a particular value. In simple terms, it tells you the chance of getting a specific result in a continuous set of possibilities.

Q2. What are the types of Probability Distribution?
   - There are several types of probability distributions, but two common ones are:
     1. Discrete Probability Distribution: This is for situations where the random variable can only take specific, distinct values (like rolling a die, where you can only get 1, 2, 3, 4, 5, or 6).
     2. Continuous Probability Distribution: This is for cases where the random variable can take any value within a range (like measuring people's heights, which can be any number within a range).

Q3. Python function to calculate the PDF of a normal distribution:
   ```python
   from scipy.stats import norm

   def calculate_normal_pdf(mean, std_dev, x):
       return norm.pdf(x, loc=mean, scale=std_dev)
   ```

Q4. Properties of Binomial distribution:
   - The Binomial distribution is used for events with two possible outcomes (success or failure) and has the following properties:
     1. Fixed number of trials (n).
     2. Constant probability of success (p).
     3. Independent trials.
   Examples: 
   a. Flipping a coin (success = heads, failure = tails).
   b. Pass/fail rates in a series of identical tests.

Q5. Generate a random sample of size 1000 from a Binomial distribution:
   ```python
   import numpy as np
   import matplotlib.pyplot as plt

   sample = np.random.binomial(n=1, p=0.4, size=1000)
   plt.hist(sample, bins=2, edgecolor='k')
   plt.xlabel("Success (1) or Failure (0)")
   plt.ylabel("Frequency")
   plt.show()
   ```

Q6. Python function to calculate the Cumulative Distribution Function (CDF) of a Poisson distribution:
   ```python
   from scipy.stats import poisson

   def calculate_poisson_cdf(mean, x):
       return poisson.cdf(x, mu=mean)
   ```

Q7. Difference between Binomial and Poisson distributions:
   - Binomial is used when there's a fixed number of trials and a constant probability of success, while Poisson is used when the number of events in a fixed interval of time or space is of interest.

Q8. Generate a random sample from a Poisson distribution and calculate mean and variance:
   ```python
   sample = np.random.poisson(lam=5, size=1000)
   sample_mean = np.mean(sample)
   sample_variance = np.var(sample)
   ```

   - In both Binomial and Poisson distributions, the mean (average) is equal to the variance (a measure of spread) when certain conditions are met. For Binomial, this happens when p (probability of success) is multiplied by n (number of trials), and for Poisson, it happens when the mean (Î») is equal to the variance.
