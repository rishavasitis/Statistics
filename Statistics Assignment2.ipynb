Q1. **Measures of Central Tendency:**
   - The three measures of central tendency are the mean, median, and mode. They help summarize the center or typical value in a dataset.

Q2. **Difference Between Mean, Median, and Mode:**
   - 1. **Mean:** It's the average of all values. It sums up all values and divides by the total count.
   - 2. **Median:** It's the middle value when data is sorted. If there's an even count, it's the average of the two middle values.
   - 3. **Mode:** It's the most frequently occurring value in the dataset.

Q3. **Measures of Central Tendency for Height Data:**
   - Mean: (Sum of values) / (Number of values) = (2693.6) / (15) = 179.5733 (approximately)
   - Median: The middle value after sorting = 178.2
   - Mode: The most frequent value = 178

Q4. **Standard Deviation for Height Data:**
   - Standard deviation measures how spread out the values are. It is approximately 1.42 (rounded to two decimal places) for the provided data.

Q5. **Measures of Dispersion to Describe Data Spread:**
   - - Range: Shows the difference between the maximum and minimum values.
     - Variance and Standard Deviation: Measure how individual data points deviate from the mean. A larger variance or standard deviation indicates greater data spread.
     - Example: For test scores, a smaller standard deviation means most students scored close to the mean, while a larger one means scores are more spread out.

Q6. **Venn Diagram:**
   - A Venn diagram is a visual tool with overlapping circles used to represent the relationships between different sets or groups. It helps show common elements and differences.

Q7. **Set Operations for A and B:**
   - (i) A ∩ B (intersection) is the set of elements that are in both A and B. Result: (2, 6)
   - (ii) A ⋃ B (union) is the set of all elements from both A and B. Result: (0, 2, 3, 4, 5, 6, 7, 8, 10)

Q8. **Skewness in Data:**
   - Skewness is a measure of the asymmetry of the data distribution. If data is skewed, it means it's not symmetric and tends to lean more to one side.

Q9. **Position of Median in Right-Skewed Data:**
   - In right-skewed data, the median is typically less than the mean. It's pulled to the right by the larger values.

Q10. **Covariance vs. Correlation:**
   - Covariance measures the direction of the linear relationship between two variables, while correlation also measures the strength of the relationship and normalizes it to a range of -1 to 1. Correlation is more informative for comparison.

Q11. **Formula for Sample Mean:**
   - Sample Mean = (Sum of all values) / (Number of values)
   - Example: For data [2, 4, 6, 8], Mean = (2 + 4 + 6 + 8) / 4 = 5

Q12. **Central Tendency in Normal Distribution:**
   - In a normal distribution, the mean, median, and mode are all approximately equal, forming a symmetric bell-shaped curve.

Q13. **Difference Between Covariance and Correlation:**
   - Covariance measures the degree of joint variability between two variables. It can be any real number.
   - Correlation standardizes the measure to be between -1 and 1, making it more interpretable and comparable.

Q14. **Effect of Outliers on Measures:**
   - Outliers can significantly impact the mean, pulling it towards extreme values. However, they have less effect on the median and mode, making these measures more robust against extreme values. For example, if you have a dataset of incomes, one person with an extremely high income can significantly inflate the mean, but it won't affect the median or mode as much.
