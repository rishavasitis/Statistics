Q1: Difference between t-test and z-test
   - A z-test and a t-test are both used for hypothesis testing, but they are used in different situations. The key difference is in the knowledge of population standard deviation:
     - Z-Test: Used when you know the population standard deviation. For example, when testing if the average weight of apples from a specific orchard is greater than a certain value.
     - T-Test: Used when you don't know the population standard deviation or when working with small samples. For instance, when comparing the average test scores of two different classes in a school.

Q2: One-Tailed vs. Two-Tailed Tests
   - One-tailed tests are used to check for a specific direction of effect (e.g., greater than or less than) in a hypothesis. For example, testing if a new medicine is better (greater than) than an existing one.
   - Two-tailed tests are more general and check for differences in both directions (e.g., not equal to) in a hypothesis. For example, checking if the new medicine has a different effect (either better or worse) than the existing one.

Q3: Type 1 and Type 2 Errors in Hypothesis Testing
   - Type 1 Error: Occurs when you mistakenly reject a true null hypothesis. An example would be a medical test wrongly indicating a healthy person has a disease.
   - Type 2 Error: Happens when you fail to reject a false null hypothesis. For instance, a medical test missing a disease in a person who actually has it.

Q4: Bayes's Theorem
   - Bayes's Theorem is a formula for finding the probability of an event, based on prior knowledge of conditions related to the event. It's used to update probabilities with new evidence. For example, you can use it to update your belief in someone's guilt based on new evidence in a criminal trial.

Q5: Confidence Interval
   - A confidence interval is a range of values used to estimate a population parameter with a certain level of confidence. To calculate it, you need the sample mean, standard deviation, sample size, and a z or t score (for a given confidence level). For example, a 95% confidence interval for the average height of students may be 160 cm to 170 cm, meaning we're 95% confident the true average falls in this range.

Q7: Calculating a 95% Confidence Interval
   - To calculate a 95% confidence interval with a sample mean of 50 and a standard deviation of 5:
     - Z-score for 95% confidence is 1.96.
     - Margin of error = 1.96 * (5 / √n), where n is the sample size.
     - If n is known (e.g., 100), the margin of error is 1.96 * (5 / √100) = 0.98.
     - The 95% confidence interval is 50 ± 0.98, or (49.02, 50.98). This means we're 95% confident the true population mean lies in this range.

Q8: Margin of Error and Sample Size
   - The margin of error is a measure of how uncertain our estimate is. It decreases as the sample size increases. In other words, a larger sample size reduces the margin of error. For example, if you increase your survey size from 100 to 1000, the margin of error decreases, making your results more precise.

Q9: Calculating Z-Score
   - The z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5 is:
     - Z = (X - Mean) / Standard Deviation = (75 - 70) / 5 = 1.
   - This z-score of 1 means that the data point is 1 standard deviation above the population mean.

Q11: Calculating Confidence Interval for Proportion
   - To calculate a 95% confidence interval for the proportion of people satisfied with their job:
     - Sample proportion = 65%
     - Z-score for 95% confidence is approximately 1.96.
     - Margin of error = 1.96 * √((0.65 * (1 - 0.65)) / 500)
     - The 95% confidence interval is 65% ± margin of error.

Q13: Calculating Confidence Interval for Population Mean
   - To calculate a 90% confidence interval for the population mean with a sample mean of 65, sample size of 50, and a standard deviation of 8:
     - Z-score for 90% confidence is approximately 1.645.
     - Margin of error = 1.645 * (8 / √50)
     - The 90% confidence interval is 65 ± margin of error.
